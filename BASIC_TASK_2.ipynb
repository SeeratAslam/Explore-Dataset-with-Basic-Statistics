# **TASK 2: Explore a Dataset with Basic Statistics**

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the dataset
from google.colab import files
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/average-monthly-surface-temperature.csv')
```

```python
df
```

```python
# Step 3: Descriptive statistics
mean_temp = df["Average surface temperature"].mean()
median_temp = df["Average surface temperature"].median()
std_temp = df["Average surface temperature"].std()
```

```python
print("Mean Temperature:", round(mean_temp, 2))
print("Median Temperature:", median_temp)
print("Standard Deviation:", round(std_temp, 2))
```

```python
# Step 4: Plot histogram
plt.hist(df["Average surface temperature"], bins=6, color="green", edgecolor="black")
plt.title("Temperature Distribution")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
